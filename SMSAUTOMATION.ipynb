```python

import os
from twilio.rest import Client

def send_sms(to_number: str, body: str) -> dict:
    
    account_sid = os.getenv("YASH")          
    auth_token  = os.getenv("AUTH_TOKEN")    
    from_number = os.getenv("FROM_NUMBER")   

    if not all([account_sid, auth_token, from_number]):
        raise EnvironmentError("Set YASH, AUTH_TOKEN, FROM_NUMBER environment variables")

    client = Client(account_sid, auth_token)

    message = client.messages.create(
        body=body,
        from_=from_number,
        to=to_number
    )
    return {"sid": message.sid, "status": message.status, "to": to_number}

if __name__ == "__main__":
    to = input("Recipient number (+91...): ").strip()
    text = input("Message: ").strip()
    result = send_sms(to, text)
    print("Sent:", result)

```
