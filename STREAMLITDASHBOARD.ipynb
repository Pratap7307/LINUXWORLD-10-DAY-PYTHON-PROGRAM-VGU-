```python
import streamlit as st
import pandas as pd


st.title("University Student Data Dashboard")


st.sidebar.header("Student Filters")
course = st.sidebar.selectbox("Select Course", ["CSE", "ECE", "ME", "CE"])


data = {
    "Name": ["Aditya", "Riya", "Mohit", "Neha", "Arjun"],
    "Roll No": [101, 102, 103, 104, 105],
    "Course": ["CSE", "ECE", "CSE", "ME", "CSE"],
    "CGPA": [8.5, 7.9, 9.1, 8.0, 7.5]
}

df = pd.DataFrame(data)


filtered_df = df[df["Course"] == course]

st.subheader("Filtered Student Data")
st.dataframe(filtered_df)


st.subheader("CGPA Chart")
st.line_chart(filtered_df["CGPA"])
```
