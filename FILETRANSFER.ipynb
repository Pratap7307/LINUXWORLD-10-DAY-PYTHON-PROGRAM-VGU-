```python
import socket
import os

def send_file(filename, host='0.0.0.0', port=5001):
    if not os.path.exists(filename):
        print(" File not found!")
        return
    
    s = socket.socket()
    s.bind((host, port))
    s.listen(1)
    print(f"ðŸ“¡ Waiting for receiver on port {port}...")

    conn, addr = s.accept()
    print(f"ðŸ”— Connected with {addr}")

   
    conn.send(os.path.basename(filename).encode())

    conn.recv(1024)

    with open(filename, "rb") as f:
        data = f.read()
        conn.sendall(data)

    print("âœ… File sent successfully!")
    conn.close()
    s.close()

```
