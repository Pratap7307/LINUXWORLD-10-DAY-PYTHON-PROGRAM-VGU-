```python
print("Unit Converter")
print("1. Km to m")
print("2. m to Km")
print("3. Kg to Gram")
print("4. Gram to Kg")

choice = input("Choose option 1/2/3/4: ")

if choice == "1":
    km = float(input("Enter value in Km: "))
    m = km * 1000
    print("meters:", m)

elif choice == "2":
    m = float(input("Enter value in Meters: "))
    km = m / 1000
    print("Km:", km)

elif choice == "3":
    kg = float(input("Enter value in Kg: "))
    g = kg * 1000
    print("Grams:", g)

elif choice == "4":
    g = float(input("Enter value in Grams: "))
    kg = g / 1000
    print("Kg:", kg)

else:
    print("Invalid Option")
```
